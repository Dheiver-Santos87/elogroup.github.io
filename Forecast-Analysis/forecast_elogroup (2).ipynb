## Análise Forecast geral

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from fbprophet import Prophet

m = Prophet()
m.fit(_prophet_df)
```

```python
# Forcasting into the future
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```
